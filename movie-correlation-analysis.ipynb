```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read the data
df = pd.read_csv(r'../input/movies/movies.csv')
```

```python
# Look at the data
df.head()
```

```python
# See if there is any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
# Drop null rows
df.dropna(inplace=True)
df.head()
```

```python
# Check the data type for each column
print(df.dtypes)
```

```python
# Change data type of 'budget' column from float to integer
df['budget'] = df['budget'].astype(int)
df['gross'] = df['gross'].astype(int)
print(df['budget'].dtypes, df['gross'].dtypes)
```

```python
# Create a correct year column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
print(df['yearcorrect'].head())
```

```python
# Set the maximum number of rows that pandas displays
pd.set_option('display.max_rows', 10)
```

```python
# Sort
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates
df.drop_duplicates()
```

```python
# Hypothesise that budget has high correlation with gross
# Hypothesise that company has high correlation with gross

# Scatter plot with budget vs gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Plot budget vs gross using seaborn
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Look at budget - the numeric correlation
#There are three methods to calculate correlation by pandas: pearson, kendall, spearman
df.corr(method='pearson')
```

```python
# High correlation between budget and gross, which justified one of the hypotheses
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Look at company - nonnumeric correlation
# Numerise companies
df_numerised = df
for col_name in df_numerised.columns:
    if (df_numerised[col_name].dtype == 'object'):
        df_numerised[col_name] = df_numerised[col_name].astype('category')
        df_numerised[col_name] = df_numerised[col_name].cat.codes
        
df_numerised
```

```python
# Look at the correlation among every attribute
correlation_matrix = df_numerised.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Look at the correlation by pairing up every attribute
correlation_mat = df_numerised.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
# Pair up attributes in ascending order of correlation index
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Filter the highest correlation pairs
high_corr = sorted_pairs[(sorted_pairs > 0.5) & (sorted_pairs < 1)]
high_corr
```
